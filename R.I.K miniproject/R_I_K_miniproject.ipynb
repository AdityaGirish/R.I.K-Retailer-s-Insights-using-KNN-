**Training and texting**

```python
import pandas as pd
import numpy as np


df = pd.read_csv("https://raw.githubusercontent.com/futurexskill/projects/main/knn-classification/purchase_history.csv")


df.head()
```

```python
df.count()
```

```python
len(df)
```

```python
gender_encoded = pd.get_dummies(df['Gender'])
```

```python
gender_encoded
```

```python
gender_encoded = pd.get_dummies(df['Gender'],drop_first=True)
```

```python
gender_encoded
```

```python
df = pd.concat([df,gender_encoded],axis=1)
```

```python
df
```

```python
x = df[['Male','Age','Salary','Price']].to_numpy() #excluded customer ID and Product ID because no significant impact
```

```python
x
```

```python
y = df['Purchased'].to_numpy()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k= 5
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
import pickle
```

```python
with open('knn_model.pickle','wb') as f:
  pickle.dump(knn,f)
```

```python
with open('scaler.pickle','wb') as f:
  pickle.dump(scaler,f)
```
